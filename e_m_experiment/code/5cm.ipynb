```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import io

fontsize=15

data = pd.read_csv(r'C:\Users\devkh\Desktop\Phys 2925\e_m_experiment\data\5cm.csv')

mu_0 = 4 * np.pi * 10**(-7) # H/m
```

```python
data.head()
```

```python
data['r (m)'] = data['r (cm)'] / 100
data['dr (m)'] = data['dr (cm)'] / 100
data.head()
```

```python
def magnetic_field(df, radius=5, N=2):
    '''Calculate the magnetic field around the Wehnelt cylinder.
    
    Parameters
    ----------
    df: pd.Dataframe containing the current measurements.
    radius: radius of the circle made by the electron beam.
    N: Number of turns around the coil.

    returns
    -------
    df: the dataframe containging the magnetic field and r^2 B^2
    '''
    k = (8 * mu_0 * N) / (np.sqrt(125) * radius)
    df['B (T)'] = k * df['I_H (A)']
    df['r^2 B^2 (m^2 T^2)'] = df['r (m)'] ** 2 + df['B (T)'] ** 2
    return df

data = magnetic_field(data, radius=5, N=2)
data
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
sns.scatterplot(
    x=2*data['V_a (V)'], 
    y=data['r^2 B^2 (m^2 T^2)'],
    c='magenta',
    ax=ax
)
plt.show()
```

```python
from scipy.optimize import curve_fit

def power_law(x, a, b):
    # y = a * x**b
    return a * np.power(x, b)

popt, pcov = curve_fit(power_law, df['Voltage (V)'], df['Current (mA)'], p0=[1, -1])
a_fit, b_fit = popt
print(a_fit)
print(b_fit)
plt.scatter(df['Voltage (V)'], df['Current (mA)'], label='Data')
plt.plot(df['Voltage (V)'], power_law(df['Voltage (V)'], a_fit, b_fit), 'r--')
```
