```python
import numpy as np
import pandas as pd
import io
# pd.set_option('display.precision', 3)
```

```python
raw_data = """
0.0000 5.2171e-06
0.050000 0.00021262
0.10000 0.00042189
0.15000 0.00062993
0.20000 0.00083599
0.25000 0.0010435
0.30000 0.0012507
0.35000 0.0014571
0.40000 0.0016656
0.45000 0.0018709
0.50000 0.0020774
0.55000 0.0022837
0.60000 0.0024899
0.65000 0.0026961
0.70000 0.0029013
0.75000 0.0031088
0.80000 0.0033163
0.85000 0.0035206
0.90000 0.0037263
0.95000 0.0039338
1.0000 0.0041370"""

# Read the data using whitespace delimiter
# If using a file, replace io.StringIO(raw_data) with 'data.txt'
df = pd.read_csv(io.StringIO(raw_data), sep=r'\s+', header=None, names=['x', 'y'])
df['x**2'] = df['x'] ** 2
df['x*y'] = df['x'] * df['y']
print(f'sum x_i = {df['x'].sum()}')
print(f'sum y_i = {df['y'].sum()}')
print(f'sum x_i^2 = {df['x**2'].sum()}')
print(f'sum x_i * y_i = {df['x*y'].sum()}')
```

```python
xi = df['x'].sum()
yi = df['y'].sum()
xi_squared = df['x**2'].sum()
xi_yi = df['x*y'].sum()
s_my_squared = 1.5e-7 # volts
```

# a_0

```python
def calculate_a0(df, xi, yi, xi_squared, xi_yi, s_my_squared):
    numerator = (xi_squared * yi) - (xi * xi_yi)
    denominator = (len(df) * xi_squared) - (xi)**2

    num = s_my_squared * xi_squared
    denom = (len(df) * xi_squared) - (xi)**2
    
    return numerator / denominator, np.sqrt(num / denom)

a_0, s_ma0 = calculate_a0(df, xi, yi, xi_squared, xi_yi, s_my_squared)

print(f'{a_0} +- {s_ma0}')
```

# a_1

```python
def calculate_a1(df, xi, yi, xi_squared, xi_yi, s_my_squared):
    numerator = (len(df) * xi_yi) - (xi * yi)
    denominator = (len(df) * xi_squared) - xi**2

    num = len(df) * s_my_squared
    denom = (len(df) * xi_squared) - xi**2

    return numerator / denominator, np.sqrt(num / denom)

a_1, s_ma1 = calculate_a1(df, xi, yi, xi_squared, xi_yi, s_my_squared) 

print(f'{a_1} +- {s_ma1}')
```
